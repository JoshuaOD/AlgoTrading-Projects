# Stock Diversification using K-Means Algorithm

```python
#import relevant packages
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.cluster import KMeans
from pandas_datareader import data as pdr
from datetime import datetime, timedelta
yf.pdr_override()
```

```python
sp500_data = pd.read_csv('sp500_timeseries.csv')
```

```python
sp500_data = sp500_data.drop(['Unnamed: 0'], axis=1)
```

```python
sp500_data
```

```python
daily_returns = sp500_data.pct_change()
annual_return = daily_returns.mean() * 252
annual_variance = daily_returns.var() * 252
```

```python
data = pd.DataFrame(data = sp500_data.columns, columns=['Tickers'])
data['Returns'] = annual_return.values
data['Variances'] = annual_variance.values
```
